```python
amount = int(input("enter an amount"))
if (amount >= 100 and amount <= 500):
    discount_1 = amount*0.10
    print("the final amount is",amount - discount_1)
elif (amount > 500):
    discount_2 = amount * 0.20
    print("the final amount is",amount - discount_2)
else:
    print("no discount")
```

```python
color = input("enter a color name: ")
color = color.lower()
if (color == "red"):
    print("STOP")
elif (color == "yellow"):
    print("Ready to move")
elif (color == "green"):
    print("Go")
else:
    print("invalid color")
```

```python
marks = int(input("enter your marks: "))
if marks >= 90:
    print("grade A")
elif marks >= 70 and marks < 90:
    print("grade B")
elif marks >= 50 and marks < 70:
    print("garde C")
else:
    print("fail")
```

```python
num = int(input("enter a value: "))
# even or odd
if num % 2 == 0:
    print("num is even")
else:
    print("num is odd")

# divisible by 5 or not
if num % 5 == 0:
    print("num is divisible by 5")
else:
    print("num is not divisible by 5")
```

```python
password = input("enter a password")
if (len(password) >= 8):
    print("strong")
elif(len(password) >= 5 and len(password) <= 7):
    print("medium")
else:
    print("weak")
```
